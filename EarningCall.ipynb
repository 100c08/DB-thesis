```python
import requests
import pandas as pd

# 제공된 API 키
API_KEY = "TO39srzvTbZb+DxgIHo4IA==f87PYAPOCcTbL3bt"

# 대상 티커 목록 (BRK-A로 수정)
tickers = ["AAPL", "AMZN", "BRK-A", "GOOGL", "META", "MSFT", "TSLA", "WMT", "AVGO", "NFLX"]

# 분석할 연도 및 분기 설정 (2024년, 2025년)
years = [2024, 2025]
quarters = [1, 2, 3, 4]

results = []

for ticker in tickers:
    for year in years:
        for quarter in quarters:
            api_url = f"https://api.api-ninjas.com/v1/earningstranscript?ticker={ticker}&year={year}&quarter={quarter}"
            response = requests.get(api_url, headers={'X-Api-Key': API_KEY})
            print(f"Processing {ticker} - {year} Q{quarter}: Status {response.status_code}")
            if response.status_code == 200:
                transcript = response.text
            else:
                transcript = f"Error {response.status_code}: {response.text}"
            results.append({
                "ticker": ticker,
                "year": year,
                "quarter": quarter,
                "transcript": transcript
            })

# DataFrame 생성 및 CSV 저장
df = pd.DataFrame(results)
df.to_csv("earnings_transcripts.csv", index=False)
print("CSV file saved as earnings_transcripts.csv")
```

```python
import yfinance as yf
import pandas as pd

# 관심 종목 리스트
tickers = ["AAPL", "AMZN", "BRK-A", "GOOGL", "META", "MSFT", "TSLA", "WMT", "AVGO", "NFLX"]

# 모든 추천 데이터를 저장할 리스트
all_recommendations = []

for ticker in tickers:
    print(f"Fetching recommendations for {ticker}...")
    tkr = yf.Ticker(ticker)
    recs = tkr.recommendations  # 추천 데이터(DataFrame) 가져오기
    if recs is not None and not recs.empty:
        recs = recs.copy()  # 원본 데이터 손상 방지
        recs["Ticker"] = ticker  # 어느 종목의 데이터인지 표시
        all_recommendations.append(recs)
    else:
        print(f"No recommendations data available for {ticker}")

# 데이터가 하나 이상 존재하면 CSV로 저장
if all_recommendations:
    final_df = pd.concat(all_recommendations)
    final_df.to_csv("analyst_recommendations.csv", index=True)
    print("Analyst recommendations data saved to 'analyst_recommendations.csv'")
else:
    print("No recommendation data found for the provided tickers.")
```

```python
from yahooquery import Ticker
import pandas as pd

# 대상 티커 목록
tickers = ["AAPL", "AMZN", "BRK-A", "GOOGL", "META", "MSFT", "TSLA", "WMT", "AVGO", "NFLX"]

all_data = []

for t in tickers:
    try:
        print(f"Processing {t}...")
        # yahooquery Ticker 객체 생성
        obj = Ticker(t)
        # recommendationTrend 속성은 각 회사의 애널리스트 추천 추세 데이터를 제공합니다.
        trend = obj.recommendationTrend
        if isinstance(trend, pd.DataFrame) and not trend.empty:
            trend['ticker'] = t
            all_data.append(trend)
        else:
            print(f"No recommendation trend data for {t}")
    except Exception as e:
        print(f"Error processing {t}: {e}")

if all_data:
    df_all = pd.concat(all_data, ignore_index=True)
    # recommendationTrend DataFrame은 'period' 열을 포함합니다.
    # 'period'가 문자열이면 datetime으로 변환
    df_all['period'] = pd.to_datetime(df_all['period'], errors='coerce')
    # 2024년과 2025년 데이터로 필터링
    df_all = df_all[df_all['period'].dt.year.isin([2024, 2025])]
    df_all.sort_values("period", inplace=True)
    output_csv = "yahoo_analyst_recommendation_trend.csv"
    df_all.to_csv(output_csv, index=False)
    print(f"CSV file saved as {output_csv}")
else:
    print("No recommendation trend data found for the selected tickers.")
```
